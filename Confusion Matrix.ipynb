```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
labels =  ['COVID19', 'LUNG CANCER', 'NORMAL', 'PNEUMONIA', 'TUBERCULOSIS']
confusion_matrix = array([[ 515  ,  0   , 4 ,   3  , 11],
                         [   0  ,137  ,  1  ,  0   , 0],
                         [   2   , 1 , 730  , 20   , 0],
                         [   3   , 0  , 24 ,1024  ,  0],
                         [  11   , 1 ,   1  ,  0 , 137]])
fig, ax = plt.subplots(figsize=(8,6), dpi=100)
display = ConfusionMatrixDisplay(confusion_matrix, display_labels=labels)
ax.set(title='Confusion Matrix for CNN Model')
display.plot(ax=ax);
```
